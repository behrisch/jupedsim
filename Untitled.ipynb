```python
import pandas as pd
from matplotlib import pyplot as plt



plt.rcParams["figure.figsize"] = [20, 10]
plt.rcParams["figure.autolayout"] = True
plt.rcParams["figure.dpi"] = 100
plt.rcParams["font.size"] = 20

data_file = "times.csv"
image_name = "py312-jps111"
title = "Apple M1 2020"

df = pd.read_csv(data_file, header=None)
df = df.rename(columns={0:"Iterations", 1:"Total time", 2:"Time spent in C++", 3:"Time spent in Python"})

ax = df.iloc[1:,2:].plot(title=f"Performance over time on {title}", xlabel="Iterations", ylabel="time[ms]", kind="area", ylim=[0,20])
ax = df.iloc[1:,0].plot(ax = ax, secondary_y = True, ylabel="#Agents", legend=True)
ax.get_figure().savefig(f"{image_name}.png")
```
